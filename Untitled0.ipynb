## Small world

* most nodes are not neighbors of one another
* most nodes can be reached from every other by a small number of steps
* The typical distance L between two randomly chosen nodes (the number of steps required) grows proportionally to the log of the number of nodes N.

$$ L \propto \log (N)$$

###Examples:

* Social networks 
* Pages in Wikipedia
* Connectivity of the Internet
* Gene networks

Centrality of a vertex is the importance of a vertex in a graph. There are 4 main measures of centrality:

1. Degree centrality: Fraction of nodes that the node in question is connected to. 
2. Clustering coefficient: A measure of the degree to which nodes cluster

![alt text](files/clustering_index.png "more text")

Closeness: Its the inverse of the mean of distances to all other nodes, this is a different metric of a clustering coefficient when we're not looking at triangles. 


![alt text](files/closeness_graph.png "more text")

![alt text](files/closeness_table.png "more text")

4. Betweenness, the number of times a node acts as a bridge along the shortest paths between 2 other nodes. 
5. Eigenvector: influence of a node in a network. It assigns relative scores based on the assumption that Connections to high scoring nodes contribute to the node-score more than connections to low scoring nodes. Google PageRank is similar.

## Scale free graphs
These graphs are characterized by a power law degree distribution. 


![alt text](files/power_vs_bell.png "Logo Title Text 1")

The rich get richer, preferential attachment.

![they are everywhere](files/power_laws_everywhere.png)



### Characteristics: 

* Nodes with many edges are hubs and usually have a functional purpose (distinct phenotype)
* Robustness to failure --> hubs attract nodes with lower connectivity which attract nodes with lower connectivity --> fault tolerant.
* If a random node is faulty, the connectivity of the graph is not severly compromised.
* Coordinated attacks: if we remove several hubs, the network collapses, there is a tradeoff. 
* Clustering coefficient decreases as the node degree increases
* The power law arises from the fraction of the nodes that have a degree k : 

$$ P(k) \sim \ k^\boldsymbol{-\gamma}, \ \ \ \ \ \  2 > \gamma > 3 $$

## Community detection
There are several algorithms to detect communities:
Detection: 

   * Girvan - Newman algorithm:  
      1. Calculate the betweenness of all existing edges in the network is calculated first 
      2. Remove the edge with the highest betweenness.
      3. Recalculate the betweenness of all edges affected by the removal of that edge.
      4. Repeat steps 2 and 3 until no edges remain.
      * Scales poorly (O(m2n) ) for m edges and n nodes. 
   * Minimum cut method
   * Hierarchical clustering
   * Modularity maximization: modularity Q: A measure of how well a network is partitioned into communities
   * Clique based methods
